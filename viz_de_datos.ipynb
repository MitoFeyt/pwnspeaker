```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Santiago':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Valparaiso':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Santiago'])
ax.plot(dias, temperaturas['Valparaiso'])
ax.set_title('Evolución de la temperatura diaria', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Paso 1: Generar datos de temperaturas
def generar_temperaturas(dias):
    temperaturas = [random.uniform(-10, 35) for _ in range(dias)]
    return temperaturas

# Paso 2: Visualizar las temperaturas
def graficar_temperaturas(temperaturas):
    dias = list(range(1, len(temperaturas) + 1))
    plt.plot(dias, temperaturas, marker='o', linestyle='-', color='b')
    plt.title('Simulación de Temperaturas Diarias en °C')
    plt.xlabel('Día del Mes')
    plt.ylabel('Temperatura (°C)')
    plt.grid(True)
    plt.show()

# Ejecución
# Generar temperaturas para 30 días (un mes)
temperaturas_mensuales = generar_temperaturas(30)

# Graficar las temperaturas generadas
graficar_temperaturas(temperaturas_mensuales)
```
